## UNIVARIATE ANALYSIS

```python
## Numerical caterogies 
```

```python
sns.histplot(taxis['color'])
plt.title('Distribution of color')
```

```python

sns.histplot(taxis['payment'])
plt.title('Modes of Payment')
```

```python
sns.histplot(taxis['fare'], kde=False)
```

```python
sns.boxplot(x=taxis['fare'])
plt.title('Boxplot of fare')
```

```python
sns.boxplot(x=taxis['distance'])
plt.title('Boxplot of distance')
plt.xticks(rotation=90)
```

```python
sns.boxplot(x=taxis['pickup'])
plt.title('Boxplot of tip')
plt.xticks(rotation=90)
```

```python
## Categorial column
```

```python
sns.countplot(x='payment',data=taxis)
```

```python
sns.countplot(x='passengers',data=taxis)
```

```python
# bivariate analysis
```

```python
sns.scatterplot(x='total',y='tip',data=taxis,hue='tip',palette='coolwarm')
plt.title('Total vs Tip')
plt.show()
```

```python
sns.scatterplot(x='passengers',y='tip',data=taxis,hue='distance',palette='coolwarm')
plt.title('passengers vs tip')
plt.show()
```

```python
taxis.corr(numeric_only=True)
```

```python
cor=taxis.corr(numeric_only=True)

sns.heatmap(cor,annot=True,cmap='coolwarm')
plt.title('correlation between numerical values')
```

```python
sns.barplot(x='passengers',y='distance',data= taxis)
plt.title('distance based on number of passengers')
```

```python
sns.violinplot(x='total',y='tip',data=taxis)
plt.title('Tip amount based on total')
```

```python
sns.lineplot(x='fare',y='tip',data=taxis)
```

```python
sns.pairplot(taxis,hue='tip')
```
