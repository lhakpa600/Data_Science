## Dataset

```python
import warnings
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 

warnings.simplefilter(action='ignore',category=FutureWarning)
df=pd.read_csv('Data/obesity.csv',index_col='Year Code')
df.head()
```

## 1. Bar plot

```python
plt.figure(figsize=(6,4))
df=pd.read_csv("Data/Obesity.csv")
sns.barplot(x='Year', y='Value', data=df)
plt.title('Prevalence of Obesity in the adult population (18 years and older) in Nepal')
plt.xlabel('Year')
plt.ylabel('%')
plt.xticks(rotation=90)
plt.show()
```

## 2. pie chart

```python
# pir chart for value 
#astype(str) was done to convert year values into strings,which is necessary for labelling process.

plt.figure(figsize=(6,6))
plt.pie(df['Value'],labels=df['Year'].astype(str), autopct='%1.1f%%', startangle=180)
plt.title('Prevalence of Obesity in the adult population in Nepal')
plt.show()


        
```

```python
df['Value'].value_counts()
```

## 3. Histogram 

```python
#assuming df['Value'] contains the data for the y-axis  

plt.figure(figsize=(10,8))
sns.histplot(df['Year'].dropna(), kde=True,color='red', label='obesity')
plt.xticks(rotation=45) #optional: Rotate x-axisticks if they overlap
plt.title('Prevalence of Obesity in the adult population in Nepal')
plt.legend()
plt.show()

```

# 4. Box plot 

```python
plt.figure(figsize=(8,6))
sns.boxplot(x='Year', y='Value', data=df, hue='Year', palette='coolwarm', legend=False)

plt.title('Prevalence of Obesity in the adult population in Nepal')
plt.xlabel('Year')
plt.ylabel('Value')
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

## 5. Violin plot 

```python
plt.figure(figsize=(8,6))

sns.violinplot(x='Year', y='Value', data=df)
plt.title('Prevalence of Obesity in the adult population in Nepal')
plt.xlabel('Year')
plt.ylabel('Value')
plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

## 6. Swarm Plot

```python
plt.figure(figsize=(10,6))
sns.stripplot(x='Year', y='Value', data=df, hue='Value', palette='bright')

plt.title('Prevalence of Obesity in the adult population in Nepal')
plt.xlabel('Year')
plt.ylabel('Value')

plt.grid(True)
plt.xticks(rotation=90)
plt.show()
```

## Heatmap for Correlation Matrix 

```python
# Import necessary libraries
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt 

# load the data 
df=pd.read_csv("Data/obesity.csv")

# check the data types of each column 
print(df.dtypes)

# drop the non numeric columns or select only numeric columns for correlation 
numeric_df=df.select_dtypes(include=['float64','int64'])

# Assuming 'df' is your DataFrame, calculate the correlation matrix
corr_matrix = numeric_df.corr()

# set the figure size for the heatmap
plt.figure(figsize=(10,8))

# use seaborn to create the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)

plt.title('Correlation Matrix of Numeric Features')

plt.show()
```
