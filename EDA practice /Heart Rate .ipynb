## Data Visualizations 

```python
import warnings
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 

## Suppress future warnings 
warnings.simplefilter(action='ignore',category=FutureWarning)

# load the Dataset 
df=pd.read_csv('Data/heart.csv',index_col='age')

# display the first few rows to verify data
df.head()
```

```python
# Barplot



x=(heart.sex.value_counts())
print(f'Number of people having sex as 1 are {x[0]} and Number of people having sex as 0 are {x[1]}')
p = sns.countplot(data=heart, x="sex")
plt.show()

```

```python
x=(heart.sex.value_counts())
print(x)
p = sns.countplot(data=heart, x="cp")
plt.show()
```

```python
x=(heart.sex.value_counts())
print(x)
p = sns.countplot(data=heart, x="thall")
plt.show()
```

```python
# PIE CHART

sex_counts = df['sex'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(sex_counts, labels=sex_counts.index, autopct='%1.1f%%',colors=['skyblue'
,
'orange'
,
'green'], startangle=180)

plt.title('Sex Distribution of Heart Attack ')
plt.show()
```

```python
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

# Suppress the specific FutureWarning
warnings.filterwarnings("ignore", category=FutureWarning)

# Your plotting code
plt.figure(figsize=(10,10))
sns.displot(heart.age, color="red", label="Age", kde=True)
plt.legend()
plt.show()
```

```python
# Your plotting code
plt.figure(figsize=(20,20))
sns.displot(heart.trtbps, color="green", label="Resting Blood Pressure", kde=True)
plt.legend()
plt.show()
```

```python
# Plot with histplot for both groups
plt.figure(figsize=(10,10))
sns.histplot(heart[heart['output'] == 0]["age"], color='green', kde=True, label='No Heart Attack')
sns.histplot(heart[heart['output'] == 1]["age"], color='red', kde=True, label='Heart Attack')
plt.title('Attack versus Age')
plt.legend()
plt.show()
```

```python
# Updated plot using histplot
plt.figure(figsize=(10,10))
sns.histplot(heart[heart['output'] == 0]["chol"], color='green', kde=True, label='No Heart Attack')
sns.histplot(heart[heart['output'] == 1]["chol"], color='red', kde=True, label='Heart Attack')
plt.title('Cholesterol versus Age')
plt.legend()
plt.show()
```

```python
# Updated plot using histplot
plt.figure(figsize=(10,10))
sns.histplot(heart[heart['output'] == 0]["trtbps"], color='green', kde=True, label='No Heart Attack')
sns.histplot(heart[heart['output'] == 1]["trtbps"], color='red', kde=True, label='Heart Attack')
plt.title('Resting Blood Pressure (trtbps) versus Age')
plt.legend()
plt.show()
```

```python
# Replace inf with NaN and drop rows with NaN values
import numpy as np
heart.replace([np.inf, -np.inf], np.nan, inplace=True)
heart.dropna(inplace=True)

# Plotting the pairplot
plt.figure(figsize=(20,20))
sns.pairplot(heart)
plt.show()
```

```python
plt.figure(figsize=(13,13))
plt.subplot(2,3,1)
sns.violinplot(x = 'sex', y = 'output', data = heart)

plt.show()
```

```python
plt.figure(figsize=(13,13))
plt.subplot(2,3,2)
sns.violinplot(x = 'thall', y = 'output', data = heart)
plt.subplot(2,3,3)
sns.violinplot(x = 'exng', y = 'output', data = heart)
plt.subplot(2,3,4)
sns.violinplot(x = 'restecg', y = 'output', data = heart)
plt.subplot(2,3,5)
sns.violinplot(x = 'cp', y = 'output', data = heart)
plt.xticks(fontsize=9, rotation=45)
plt.subplot(2,3,6)
sns.violinplot(x = 'fbs', y = 'output', data = heart)

plt.show()
```
